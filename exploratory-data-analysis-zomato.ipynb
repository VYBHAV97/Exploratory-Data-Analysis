```python
#Identify missing values using list comprehension
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
#Using Heat map
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10,5)
sns.heatmap(df.isnull(),yticklabels = False,cbar=False,cmap = 'viridis')
```

```python
df_country = pd.read_excel("/kaggle/input/zomato-restaurants-data/Country-Code.xlsx")
```

```python
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on = 'Country Code', how = 'left')
```

```python
final_df.head(10)
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
#Plot pie chart for top 3 conntries
plt.pie(country_values[:3],labels = country_names[:3],autopct = "%1.2f%%")
```

Observation : Zomato maximum records or transactions are from India and then USA and UK

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

# Observation :
1.When rating is between 4.5 to 4.9 ----> Excellent
2.When rating is between 4.0 to 3.4 ----> Very Good
3.When raing is bewtween 3.5 to 3.9 ----> good
4.When rating is between 3.0 to 3.4 ----> average
5.When rating is between 2.5 to 2.9 ----> average
4.When rating is between 2.0 to 2.4 ----> Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (14,8)
sns.barplot(x = 'Aggregate rating',y = 'Rating Count',data = ratings)
```

```python
sns.barplot(x = 'Aggregate rating',y = 'Rating Count',data = ratings,hue = 'Rating color',palette = ['blue','red','orange','yellow','green','green'])
```

Observation

Not Rated - count is very high
Maximum number of ratings are between 2.5 to 3.5

```python
#Count Plot
sns.countplot(x = "Rating color",data = ratings,palette = ['blue','red','orange','yellow','green','green'])
```

```python
#Country names that has given 0 rating
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
```

Observations

Maximum number of 0 ratings are from Indian customers

```python
#Currency for respective country?
final_df.groupby(['Country','Currency']).size().reset_index()
```

```python
#Countries with online delivery options
final_df[final_df['Has Online delivery']=='Yes'].Country.value_counts()
```

```python
#Countries with and without online delivery options
final_df.groupby(['Country','Has Online delivery']).size().reset_index()
```

Observations

Online deliveries are available in India and UAE
